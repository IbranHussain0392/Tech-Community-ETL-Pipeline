```python
with open("finaldata.txt",encoding='utf-8') as f:
    data = f.read()

```

```python
chunks = data.split("\n\n")
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')
    username =  sep_chunk[0]
    no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",","").replace(" ",""))
    no_of_followers = float(sep_chunk[2].split(" follower")[0].replace(",","").replace(" ","").replace("K","").replace("M",""))
    if(" K" in sep_chunk[2]):
         no_of_followers =  int(no_of_followers *1000)
    elif(" M" in sep_chunk[2]):
         no_of_followers = int( no_of_followers *1000000)
    else:
         no_of_followers = int( no_of_followers)
        
    no_of_following = float(sep_chunk[3].split(" following")[0].replace(",","").replace(" ","").replace("K","").replace("M",""))
    if(" K" in sep_chunk[3]):
         no_of_following =  int(no_of_following *1000)
    elif(" M" in sep_chunk[3]):
         no_of_following = int( no_of_following *1000000)
    else:
         no_of_following= int( no_of_following)
        
    name = sep_chunk[4]
    type_of_page =  sep_chunk[5]
    bio = "\n".join(sep_chunk[6:])
    return { "username":username, "no_of_posts":no_of_posts, "no_of_followers":no_of_followers, "no_of_following":no_of_following,
            "name":name, "type_of_page":type_of_page, "bio":bio}
```

```python
all_chunks = []
for chunk in chunks:
    lines = chunk.strip().split('\n')
    if len(lines) <= 5:
        continue
    data= parse_chunk(chunk)
    all_chunks.append(data)
```

```python
import json
s = json.dumps(all_chunks,indent = 4)
with open("data.json","w") as f:
    f.write(s)
```
