```python
with open("finaldata.txt",encoding='utf-8') as f:
    data = f.read()

```

```python
data

```

```python
chunks = data.split("\n\n")
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')
    username =  sep_chunk[0]
    no_of_posts = int(sep_chunk[1].split(" post")[0].replace(",","").replace(" ",""))
    no_of_followers = float(sep_chunk[2].split(" follower")[0].replace(",","").replace(" ","").replace("K","").replace("M",""))
    if(" K" in sep_chunk[2]):
         no_of_followers =  int(no_of_followers *1000)
    elif(" M" in sep_chunk[2]):
         no_of_followers = int( no_of_followers *1000000)
    else:
         no_of_followers = int( no_of_followers)
        
    no_of_following = float(sep_chunk[3].split(" following")[0].replace(",","").replace(" ","").replace("K","").replace("M",""))
    if(" K" in sep_chunk[3]):
         no_of_following =  int(no_of_following *1000)
    elif(" M" in sep_chunk[3]):
         no_of_following = int( no_of_following *1000000)
    else:
         no_of_following= int( no_of_following)
        
    name = sep_chunk[4]
    type_of_page =  sep_chunk[5]
    bio = "\n".join(sep_chunk[6:])
    return { "username":username, "no_of_posts":no_of_posts, "no_of_followers":no_of_followers, "no_of_following":no_of_following,
            "name":name, "type_of_page":type_of_page, "bio":bio}
    
#print(parse_chunk(chunks[0]))
```

```python
all_chunks = []
for chunk in chunks:
    lines = chunk.strip().split('\n')
    if len(lines) <= 5:
        continue
    data= parse_chunk(chunk)
    all_chunks.append(data)
print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks,indent = 4)
with open("data.json","w") as f:
    f.write(s)
print(s)
```

# who has the maximum posts?

lets write code to find max posts


```python
import matplotlib.pyplot as plt
max_posts = 0
for chunk in all_chunks:
    if(max_posts<chunk['no_of_posts']):
        max_posts= chunk['no_of_posts']
        chunk_with_max_posts = chunk
print(chunk_with_max_posts)
chunk_names = [f"chunk{i+1}" for i in range(len(all_chunks))]
post_counts = [chunk['no_of_posts'] for chunk in all_chunks]
colors = ['orange' if count == max_posts else 'skyblue' for count in post_counts]
plt.figure(figsize=(15,7))
plt.bar(chunk_names,post_counts,color=colors,edgecolor="black")
plt.xlabel("chunks",fontsize=12)
plt.xticks(rotation=90,fontsize=9)
plt.ylabel("Number Of Post",fontsize=12)
plt.title("Distribution Of Post Across chunks",fontsize=14)
plt.tight_layout()
plt.show()
    
```

# How many categories?
lets write code to find out number of categories

```python
categories = set()
for chunk in all_chunks:
    categories.add(chunk['type_of_page'])
print(categories,f"There are { len(categories)} categories")
```

```python
import matplotlib.pyplot as plt

# Categories ka count nikalna
cat_counts = {}
for chunk in all_chunks:
    cat = chunk['type_of_page']
    cat_counts[cat] = cat_counts.get(cat, 0) + 1

# Data ko sort karna (Highest upar)
sorted_cats = sorted(cat_counts.items(), key=lambda x: x[1], reverse=True)
names, counts = zip(*sorted_cats)

plt.figure(figsize=(12, 10))
plt.barh(names, counts, color='skyblue', edgecolor='black')
plt.xlabel('Number of Pages')
plt.title('Category-wise Distribution(Clear View)')
plt.gca().invert_yaxis() # Highest category ko top par dikhane ke liye
plt.tight_layout()
plt.show()
```
